## 双目定位基本原理
---

### **一、核心原理：视差与三角测量**
1. **视差计算**  
   两相机（基线距为 **B**）从不同角度拍摄同一物体时，物体在左右图像中的水平像素位置差称为视差（Disparity）。**视差与物体到相机的距离成反比**：物体越近，视差越大；物体越远，视差越小。

2. **三角测量公式**  
   已知相机焦距 **f** 和基线距 **B**，通过视差可计算物体深度 **Z**（距离）：

   
    Z = B*f/Disparity

   例如，基线距为 10 cm，焦距为 500 ，视差为 50 像素时，物体距离为 Z=10×500/50=100cm

---

### **二、关键步骤**
1. **相机标定**  
   确定相机的**内部参数**（如焦距、畸变系数）和**外部参数**（两相机的相对位置与姿态），确保成像模型准确。标定通常使用棋盘格等参照物完成。

2. **图像校正**  
   通过极线校正技术，将左右图像的对应点约束在同一水平线上，简化匹配搜索过程。例如，非平行相机会通过单应性矩阵投影到同一平面，使极线水平对齐。

3. **特征匹配**  
   在左右图像中寻找同一物体的特征点（如边缘、角点），常用算法包括块匹配（Block Matching）或特征点匹配（如 SIFT、ORB）。匹配精度直接影响深度计算的准确性。


---

### **四、误差来源与限制**
1. **主要误差来源**：
- **安装误差**：两相机的基线距或角度偏差；
- **标定误差**：镜头畸变校正不完善；
- **特征点误差**：如PSVR手柄的光带圆角导致特征点提取模糊；
- **像素误差**：图像分辨率有限，导致视差计算离散化。

2. **应用限制**：
- **距离限制**：有效范围通常在 **2米以内**，远距离物体因视差过小导致精度骤降；
- **环境依赖**：需场景具备足够纹理，纯色墙面或重复图案会导致匹配失败；
- **光照敏感**：光线变化可能影响图像匹配的稳定性。

---

### **五、典型应用场景**
1. **工业检测**  
   芯片定位、零件尺寸测量（如网页1提到的Microvision系统），精度可达毫米级。
2. **机器人导航**  
   机械臂抓取、AGV避障，依赖实时三维坐标反馈。
3. **虚拟现实（VR）**  
   PSVR等设备通过双目摄像头追踪手柄光带实现6自由度定位。
4. **自动驾驶**  
   短距离障碍物检测（需结合激光雷达等其他传感器）。

---


## airsim中的深度相机

无人机摄像头定位，由多模态大模型+无人机自身的深度视觉相机给出。

其中多模态大模型给出所有物体和对应的bbox，深度相机则根据bbox和对应的深度信息，给出具体的定位信息。

airsim作为一个模拟仿真系统，其视觉定位并没有特定为双目定位，而是通过虚幻引擎的渲染管线直接生成像素级深度图，无需依赖传统双目匹配或结构光模拟。

```python

#airsim深度相机设置如下

{
    "SettingsVersion": 1.2,
    "SimMode": "Multirotor",
    "CameraDefaults": {
        "CaptureSettings": [
            {
                "ImageType": 0,
                "Width": 1280,
                "Height": 720,
                "FOV_Degrees": 90
            },
            {
                "ImageType": 1,
                "Width": 1280,
                "Height": 720,
                "FOV_Degrees": 90
            },
            {
                "ImageType": 2,
                "Width": 1280,
                "Height": 720,
                "FOV_Degrees": 90
            }
        ]
    }
}

```

上面设置中，其中CaptureSettings的ImageType的0,1,2，分别代表 彩色图/Scene/一般照片。深度图DepthPlanar/像素值代表到相平面的距离，深度图DepthPerspective/像素值代表透视投影下到相平面的距离

两个深度图的距离表示如下：

````python
from airsim_smol_wrapper import *

#起飞
takeoff()
````

````python
"""
获得前置摄像头渲染图像
"""
import sys
sys.path.append('../external-libraries')
import airsim
client = airsim.MultirotorClient()#run in some machine of airsim,otherwise,set ip="" of airsim
````

````python
#部分集成显卡，运行可能会比较慢。
responses = client.simGetImages([
    airsim.ImageRequest("0", airsim.ImageType.DepthPerspective, pixels_as_float=True)
])
depth_data = responses[0].image_data_float  # 浮点数组
````

````python
1280*720
````

````python
len(depth_data)
````

````python
import numpy as np
depth_array = np.array(depth_data).reshape(responses[0].height, responses[0].width)
depth_normalized = (depth_array / depth_array.max() * 255).astype(np.uint8)

import matplotlib.pyplot as plt
# 设置图像尺寸和内联显示
%matplotlib inline
plt.figure(figsize=(10, 8))  # 调整图像尺寸

# 显示灰度图，关闭坐标轴
plt.imshow(depth_normalized, cmap='gray', aspect='auto')
plt.axis('off')  # 隐藏坐标轴
plt.title("Normalized Depth Map")
plt.show()
````

````python
#部分显卡，运行速度可能比较慢
responses = client.simGetImages([
    # png format
    airsim.ImageRequest(0, airsim.ImageType.Scene, pixels_as_float=False, compress=True),

    # floating point uncompressed image，深度图, 像素点代表到相平面距离
    airsim.ImageRequest(0, airsim.ImageType.DepthPlanar, pixels_as_float=True),

    # 像素点代表的到相机的距离
    airsim.ImageRequest(0, airsim.ImageType.DepthPerspective, pixels_as_float=True)
  ]
)
````

````python
img_depth_planar = np.array(responses[1].image_data_float).reshape(responses[1].height, responses[0].width)
img_depth_perspective = np.array(responses[2].image_data_float).reshape(responses[2].height, responses[1].width)

#一般图片
image_data = responses[0].image_data_uint8
img = cv2.imdecode(np.array(bytearray(image_data), dtype='uint8'), cv2.IMREAD_UNCHANGED)  # 从二进制图片数据中读
img = cv2.cvtColor(img, cv2.COLOR_RGBA2RGB)  # 4通道转3
````

````python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# RGB图像
axes[0].imshow(img)
axes[0].set_title("RGB Scene")
axes[0].axis('off')

# 平面深度图（DepthPlanar）
planar_plot = axes[1].imshow(img_depth_planar, cmap='jet')
plt.colorbar(planar_plot, ax=axes[1], fraction=0.046, pad=0.04)
axes[1].set_title("DepthPlanar (mm)")
axes[1].axis('off')

# 透视深度图（DepthPerspective）
perspective_plot = axes[2].imshow(img_depth_perspective, cmap='viridis')
plt.colorbar(perspective_plot, ax=axes[2], fraction=0.046, pad=0.04)
axes[2].set_title("DepthPerspective (mm)")
axes[2].axis('off')

plt.tight_layout()
plt.show()
````

````python
#叠加图片
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# DepthPlanar叠加
axes[0].imshow(img)
overlay_planar = axes[0].imshow(img_depth_planar, cmap='jet', alpha=0.5)
plt.colorbar(overlay_planar, ax=axes[0], label='Distance (mm)')
axes[0].set_title("RGB + DepthPlanar Overlay")
axes[0].axis('off')

# DepthPerspective叠加
axes[1].imshow(img)
overlay_perspective = axes[1].imshow(img_depth_perspective, cmap='viridis', alpha=0.5)
plt.colorbar(overlay_perspective, ax=axes[1], label='Distance (mm)')
axes[1].set_title("RGB + DepthPerspective Overlay")
axes[1].axis('off')

plt.tight_layout()
plt.show()
````
